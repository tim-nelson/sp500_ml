# 1. Regression for Trading

### Approach

1. Extract open, close and volume data for SP500 stocks.
2. Create indicators: 
    * EMA: https://www.investopedia.com/terms/e/ema.asp  
    * MACD: https://www.investopedia.com/terms/m/macd.asp  
    * RSI: https://www.investopedia.com/terms/r/rsi.asp  
    * Bollinger Bands: https://www.investopedia.com/terms/b/bollingerbands.asp
    * % Change of each indicator
3. Apply a regression model (from sklearn) and evaluate it by comparing model trading decisions to equally investing in all stocks.

## Set-up

```python
from datetime import datetime, timedelta

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import yfinance as yf
from sklearn.linear_model import LinearRegression

sns.set(style="whitegrid")
```

```python
tickers = pd.read_html("https://en.wikipedia.org/wiki/List_of_S%26P_500_companies")[0][
    "Symbol"
]
print(tickers.head())
```

## Functions

### Data extraction/transformation

```python
## Extract stock data
def _data_setup_trade(ticker, period="15y", interval="1d"):
    data = yf.download(
        tickers=ticker, period=period, interval=interval, prepost=False
    )  # .reset_index()
    # data = yf.download(tickers=ticker, period=period, interval=interval, prepost=False).reset_index()
    data = _transform_stock_data(data)
    data = _add_indicators_trade(data)
    return data


## Transform stock data
def _transform_stock_data(raw_data: pd.DataFrame):
    data = (
        raw_data[["Open", "Close", "Volume"]]
        .copy()
        .rename(columns={"Open": "open", "Close": "close", "Volume": "volume"})
        .rename_axis("time")
    )
    return data
```

### Indicators calculations

```python
## Moving Average Convergence Divergence (MACD or MAC-D): Takes 2 exponential moving averages,
## calculates difference to see how far they are off from eachother
def _calc_macd(data, len_short, len_long, len_signal):

    ## Two EMAs
    shortEMA = data.ewm(span=len_short, adjust=False).mean()
    longEMA = data.ewm(span=len_long, adjust=False).mean()

    ## MACD
    MACD = shortEMA - longEMA

    ## Signal = EMA of MACD
    signal = MACD.ewm(span=len_signal, adjust=False).mean()

    return MACD, signal


## Relative Strength Index (RSI): whether a stock is overbought or undersold
def _calc_rsi(data, period):

    ## Difference between the price today and yesterday
    delta = data.diff()  # how much each row increases from the previous row.

    ## How often above 0, how often below 0
    up = delta.clip(lower=0)  # set all values < 0 to 0
    down = -1 * delta.clip(upper=0)  # set all values > 0 to 0, then multiply by -1

    ## EMA
    ema_up = up.ewm(com=period, adjust=False).mean()
    ema_down = down.ewm(com=period, adjust=False).mean()

    ## Relative strength number
    rs = ema_up / ema_down

    ## Relative strength index
    rsi = 100 - (100 / (1 + rs))

    return rsi


## Bollinger bands: put lower and upper band around price: 2 standard deviations away from the mean
def _calc_bollinger(data, period):
    mean = data.rolling(period).mean()  # simple moving average
    std = data.rolling(period).std()  # simple moving std
    upper_band = np.array(mean) + 2 * np.array(std)
    lower_band = np.array(mean) - 2 * np.array(std)

    return upper_band, lower_band
```

### Adding indicator columns to data

```python
def _add_indicators_trade(data):
    ## Day of week
    data["weekday"] = data.index.isocalendar().day

    ## Previous day values
    data["prev_close"] = data["close"].shift(1)
    data["prev_volume"] = data["volume"].shift(1)
    # data.drop(["close", "volume"], axis=1, inplace=True)

    ## SMA columns
    for i in (5, 10, 20, 50, 100, 200):
        data[str(i) + "sma"] = data["prev_close"].rolling(i).mean()

    ## Bollinger columns
    upper, lower = _calc_bollinger(data["prev_close"], 20)
    data["upper_band"] = upper
    data["lower_band"] = lower

    ## MACD and Signal columns
    MACD, signal = _calc_macd(data["prev_close"], 12, 26, 9)  # 12 and 26 are standard
    data["macd"] = MACD
    data["macd_signal"] = signal
    # data["macd_signal_diff"] = MACD - signal

    ## RSI column
    data["rsi"] = _calc_rsi(data["prev_close"], 13)
    data["rsi_volume"] = _calc_rsi(data["prev_volume"], 13)

    ## Compute the % change from the immediately previous row for each indicator column
    labels = data.columns[4:]
    for period in [1, 2, 5, 10]:
        new_labels = [str(period) + "d_" + label for label in labels]
        data[new_labels] = data[labels].pct_change(
            periods=period, fill_method="ffill"
        )  # ffill equiv. to pad: propagate last valid observation forward to next valid backfill (How to handle NAs before computing percent changes)

    ## Drop NaN values
    data = data.replace(np.inf, np.nan).dropna()
    # data = data.reset_index()

    return data
```

## Regression Model and Evaluation

```python
def reg_model(data):
    y = data["close"]
    X = data.drop(
        ["close", "volume"], axis=1
    ).values  # axis=1 since dropping column, .values to get numpy array

    # For time series data, do NOT use train_test_split since easy to interpolate around points that we take out.
    # Instead, cut off at a point and predict everything in the future.

    ## Split into train and test data
    num_test = 365  # last year will be test date
    test_index = data[-1 * num_test :].index
    X_train = X[: -1 * num_test]
    y_train = y[: -1 * num_test]
    X_test = X[-1 * num_test :]
    y_test = y[-1 * num_test :]

    ## Fit regresion model
    model = LinearRegression()
    model = model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    # fig, ax = plt.subplots(figsize=(30,5))
    # sns.lineplot(x=test_index, y=y_test, color="blue")
    # sns.lineplot(x=test_index, y=y_pred, color="red");

    return test_model(X_test.T[0], y_test, y_pred, 1000, 0.001)


def test_model(opens, closes, y_pred, start_balance=1000, thresh=10):
    trade_account = start_balance
    trade_changes = []

    for i in range(len(y_pred)):
        if (y_pred[i] - opens[i]) / opens[
            i
        ] >= thresh:  # predicted precent change greater than threshhold
            trade_account += (
                trade_account * (closes[i] - opens[i]) / opens[i]
            )  # add percent change into account
        trade_changes.append(trade_account)

    trade_changes = np.array(trade_changes)  # convert to numpy array for ease

    ## Plot changes
    # fig, ax = plt.subplots(figsize=(30,5))
    # sns.lineplot(x=range(len(trade_changes)), y=trade_changes)
    # plt.plot(trade_changes)

    ## Amount if we invest for a year
    invest_total = (
        start_balance + start_balance * (closes[-1] - opens[0]) / opens[0]
    )  # percent change = (last closing value - first opening value)/first opening value

    invest_percent_change = round(
        100 * (invest_total - start_balance) / start_balance, 1
    )
    trade_percent_change = round(
        100 * (trade_account - start_balance) / start_balance, 1
    )

    # print("Invest Total: ", invest_total, invest_percent_change)
    # print("Algo-Trading Total: ", trade_account, trade_percent_change)
    return {
        "Invest": invest_total,
        "Invest %": invest_percent_change,
        "Trade": trade_account,
        "Trade %": trade_percent_change,
    }
```

```python
invest_trade_df = pd.DataFrame(
    columns=["Ticker", "Invest", "Invest %", "Trade", "Trade %"]
)

for ticker in tickers[0:20]:
    data = _data_setup_trade(ticker)
    # display(data)
    # plot_data(data)
    totals = reg_model(data)
    total_row = {"Ticker": ticker}
    total_row["Invest"] = totals["Invest"]
    total_row["Invest %"] = totals["Invest %"]
    total_row["Trade"] = totals["Trade"]
    total_row["Trade %"] = totals["Trade %"]
    invest_trade_df = pd.concat(
        [invest_trade_df, pd.DataFrame(data=total_row, index=[0])],
        axis=0,
        ignore_index=True,
    )
```

# 2. Classification for Investing

### Approach

1. Extract yearly financial information SP500 stocks.
2. Create financial ratio indicators. (Extracting all the data takes a while, so I have attached a copy.) Some of the ratios are on the following cheat sheet: https://cheatography.com/reccur/cheat-sheets/financial-ratios/
3. Apply a classification model (from sklearn) and evaluate it by comparing model trading decisions to equally investing in all stocks.

## Set-up

```python
import time
from datetime import datetime

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import yfinance as yf

sns.set(style="whitegrid")
```

```python
tickers = pd.read_html("https://en.wikipedia.org/wiki/List_of_S%26P_500_companies")[0][
    "Symbol"
]
print(tickers)
```

## Functions

```python
def _data_setup_invest(tickers):
    """Data extraction and set-up; uses three functions defined below."""
    # yearly_combined = pd.DataFrame()

    for ticker in tickers:
        try:
            yearly_combined = pd.read_csv("combined.csv")

        except:
            yearly_combined = pd.DataFrame(columns=["Ticker"])

        if (
            yearly_combined["Ticker"].str.contains(rf"(?:\s|^){ticker}(?:\s|$)").sum()
            > 0
        ):
            # print(f"{ticker} already in DataFrame.")
            pass
        else:
            ## Start timer
            start = time.time()
            ## Data set-up for ticker
            try:  # try as some stocks do not have information required for all ratios
                yearly = _extract_stock_info_add_indicators(ticker)
                yearly = _remove_unecessary_cols(yearly)

                # Add data for ticker to yearly_combined
                if yearly_combined.empty:
                    yearly_combined = yearly
                else:
                    yearly_combined = pd.concat([yearly_combined, yearly]).reset_index(
                        drop=True
                    )

            except Exception as e:  # take whatever error was, save as e
                print(ticker, e)  # print ticker and related error

            ## Pull stock every 10 seconds to prevent being throttled
            time_diff = time.time() - start
            if time_diff < 10:
                time.sleep(10 - time_diff)

            yearly_combined.to_csv("combined.csv", index=False)

        # return yearly_combined


def _extract_stock_info_add_indicators(ticker):
    """Extract financial yearly data and target column for a ticker, then add indicators."""
    ## Aquiring Info and Data
    stock = yf.Ticker(ticker)
    info = (
        stock.info
    )  # pull company info, including lots of features for current year. Need to build historical data ourselves.
    balance = stock.balance_sheet.iloc[
        :, ::-1
    ]  # pull yearly balance sheet, reverse so columns in same order after transposing earnings
    balance_q = stock.quarterly_balance_sheet.iloc[
        :, ::-1
    ]  # pull quarterly balance sheet
    financials = stock.financials.iloc[:, ::-1]  # more revenue and income values
    earnings = (
        stock.earnings
    )  # revenue and earnings for up to past 4 years, reverse so columns in same order after transposing earnings
    cashflow = stock.cashflow.iloc[:, ::-1]  # cashflow

    start_dates = (
        balance.columns.values
    )  # from these dates, want to predict the year after
    end_dates = start_dates + np.timedelta64(
        366, "D"
    )  # want first daty of  year to last day of next year: add numpy 366 (D)ays

    start_dates = [str(date)[:10] for date in start_dates]
    end_dates = [str(date)[:10] for date in end_dates]

    # Make ALL column names the same
    columns = (
        earnings.T.columns.values
    )  # Earnings years (rows), transpose to columns, extract.

    balance.columns = columns
    financials.columns = columns
    earnings = earnings.T  # transpose so years are columns
    cashflow.columns = columns

    # Combine into single DataFrame
    yearly = pd.concat([balance, financials, cashflow, earnings]).T
    columns = yearly.columns.values

    # Add extra info
    yearly["Ticker"] = ticker
    yearly["Year"] = yearly.index.values
    yearly["Sector"] = info["sector"]

    ## Create Target Column
    # Create dictionary with stock history data for each year
    hist = {}
    for x in range(4):  # numbers 1 to 4
        hist["year{}".format(x + 1)] = stock.history(
            start=start_dates[x], end=end_dates[x], period="day"
        ).reset_index(
            drop=True
        )  # pull historical stock data. Index based on date, so reset to index from 0.

    # What are we trying to maximise? Return on investement. We want percent change.
    # Baseline: Price % change over each year
    start_price = np.array(
        [hist["year{}".format(x + 1)].loc[:, "Close"][[1]] for x in range(4)]
    )  # last date of year before is 0, first date of year is 1
    end_price = np.array(
        [hist["year{}".format(x + 1)].loc[:, "Close"][-1:] for x in range(4)]
    )  # last day
    # Percent change in price over year
    pct_diff = (end_price - start_price) / start_price
    yearly["Result"] = pct_diff

    ## Liquidity (used to assess the short-term solvency position of a firm i.e. firm's ability to pay short term obligations out of current/liquid assets.)
    # Current Ratio (CR) = current assets / current liabilities
    yearly["Current_Ratio"] = yearly["Current Assets"] / yearly["Current Liabilities"]

    # Quick Ratio (Acid Test Ratio) = (current assets - inventory) / current liabilities
    yearly["Quick_Ratio"] = yearly["Current Assets"]
    if "Inventory" in columns:
        yearly["Quick_Ratio"] -= yearly["Inventory"]  # not all have inventory listed
    yearly["Quick_Ratio"] /= yearly["Current Liabilities"]

    # Cash Ratio = (cash + cash equivalents) / current liabilities
    yearly["Cash_Ratio"] = yearly["Cash And Cash Equivalents"]
    if "Cash Cash Equivalents And Short Term Investments" in columns:
        yearly["Cash_Ratio"] = yearly[
            "Cash Cash Equivalents And Short Term Investments"
        ]  # not all have inventory listed
    yearly["Cash_Ratio"] /= yearly["Current Liabilities"]

    # Operating Cash Flow Ratio (OCFR) = operating cash flow / current liabilities
    yearly["Operating_Cash_Flow_Ratio"] = (
        yearly["Cash Flow From Continuing Operating Activities"]
        / yearly["Current Liabilities"]
    )

    # Equity ratio = total equity / total assets
    yearly["Equity_Ratio"] = yearly["Stockholders Equity"] / yearly["Total Assets"]

    ## Leverage/Debt Management Ratios (indicate the extent to which debt financing is used by a firm. These ratios measure long-term solvency of a firm.)
    # Debt Equity Ratio (D/E Ratio) = total liabilities (long-term debt) / shareholder equity
    yearly["Debt_Equity_Ratio"] = (
        yearly["Long Term Debt"] / yearly["Stockholders Equity"]
    )

    # Liabilities to Assets Ratio (LAR) = total liabilities / total assets
    yearly["Liabilities_to_Assets_Ratio"] = (
        yearly["Total Liabilities Net Minority Interest"] / yearly["Total Assets"]
    )

    # Interest Coverage Ratio (ICR) = operating income / interest expenses
    yearly["Interest_Coverage_Ratio"] = (
        yearly["Operating Income"] / yearly["Interest Expense"]
    )

    ## Assets Management Ratios (measure the effectiveness of a firm's asset utilization. Also Called Turnover Ratios or Efficiency or Performance Ratios because they indicate the speed with which assets are being converted into sales)
    # Total Assets Turnover Ratio (TATOR) = revenue / total assets
    yearly["Asset_Turnover_Ratio"] = yearly["Revenue"] / yearly["Total Assets"]

    # Receivables Turnover Ratio (RTOR) = revenue / accounts receivable
    if "Receivables" in columns:  # not in all of the datasets
        yearly["Receivables_Turnover_Ratio"] = yearly["Revenue"] / yearly["Receivables"]
    else:
        yearly["Receivables_Turnover_Ratio"] = np.nan

    ## Profitability
    # Gross Profit Margin (GPM) = gross profit / revenue
    yearly["Gross_Profit_Margin"] = yearly["Gross Profit"] / yearly["Revenue"]

    # Net Profit Margin (NPM) = net income / revenue
    yearly["Net_Profit_Margin"] = yearly["Net Income"] / yearly["Revenue"]

    # Operating Margin (Return on Sales, ROS) = operating income / revenue
    yearly["Operating_Margin"] = yearly["Operating Income"] / yearly["Revenue"]

    # Return on Equity (ROE) = net income / shareholder equity
    yearly["Return_On_Equity"] = yearly["Net Income"] / yearly["Stockholders Equity"]

    # Return on Assets (ROA) = net income / total assets
    yearly["Return_On_Assets"] = yearly["Net Income"] / yearly["Total Assets"]

    # Earnings Per Share (EPS) = earning / total shares outstanding
    # Need to come up with a price year: At the time of release of these documents, need to know the current share price
    if "Net Common Stock Issuance" in columns:  # not everyone reports in the same way
        yearly["EPS"] = yearly["Earnings"] / yearly["Net Common Stock Issuance"]

    # Price-Earnings Ratio (PER) = share price / EPS
    shareprice = np.array(
        [hist["year{}".format(4 - x)].loc[0, "Close"] for x in range(4)]
    )  # Note, yearsx from release of report to year after, so want .loc[0]... (year1 = 2022, year2 = 2021, year3 = 2020, year4 = 2019)
    yearly["PER"] = shareprice / yearly["EPS"]

    return yearly


def _remove_unecessary_cols(yearly):
    """Remove all columns we no longer need / keep only columns we needs"""
    columns = list(yearly.columns[-21:])  # lsit of all useful columns
    columns.pop(columns.index("Result"))  # pop Result
    columns = columns + ["Result"]  # add result column to end
    yearly = yearly[columns].reset_index(
        drop=True
    )  # reset_index resets to new index and turns index into column. drop = True, prevents index from being added as a column

    return yearly
```

## Data Pre-Processing

```python
data = pd.read_csv("combined.csv")  # "sp500_data.csv"
data.isna().sum()

## Convert Sector into categorical data
data["Sector"] = (
    data["Sector"].astype("category").cat.codes
)  # convert into categorical data, then puts integer value for each category

## Drop any rows with a missing result
data.dropna(
    subset=["Result"], inplace=True
)  # inplace=True modifies DataFrame instead of creating a new one
# data.isna().sum()

## Result is percentage change in price across year.
## Ideally, we really want to predict tiers of percent change. E.g. 1) any negative number, 2) 0-20% return, 3) 20-50% return, 4) 50+% return

bins = [-10, 0, 0.25, 0.5, 10]
sns.histplot(data=data, x="Result", bins=bins)
data["result_bins"] = (
    pd.cut(data["Result"], bins).astype("category").cat.codes
)  # cut puts into bins

## Store "Result" (target) for later and drop from DataFrame
pct_change = data["Result"]
data = data.drop(["Result"], axis=1)
```

```python
## Split out data (before taking care of missing values, to avoid carrying over any biases)
X = data[data.columns[2:-1]]  # ignore "ticker" and "result_bins"
y = data["result_bins"]

## Split train and test: (Can't consider >=2020 because there are 2021 values but we don't know their final values as we are currently mid 2022, so no 2022 report)
train_idx = data[
    data["Year"] < 2020
].index.values  # all index values assossiated with points before 2020
test_idx = data[data["Year"] == 2020].index.values  # 2020 data to predict 2021 report.

X_train = X.loc[train_idx]
X_test = X.loc[test_idx]
y_train = y.loc[train_idx]
y_test = y.loc[test_idx]


## Replace outliers with upper and lower limits
IQR = X_train.quantile(0.75) - X_train.quantile(0.25)
lower = X_train.quantile(0.25) - 1.5 * IQR
upper = X_train.quantile(0.75) + 1.5 * IQR

for i in X_train.columns:
    X_train.loc[X_train[i] < lower[i], i] = lower[
        i
    ]  # wherever current col < lower, take column and set to lower limit
    X_train.loc[X_train[i] > upper[i], i] = upper[
        i
    ]  # wherever current col > greater, take column and set to greater limit

    X_test.loc[X_test[i] < lower[i], i] = lower[
        i
    ]  # wherever current col < lower, take column and set to lower limit
    X_test.loc[X_test[i] > upper[i], i] = upper[
        i
    ]  # wherever current col > greater, take column and set to greater limit

    ## Fill in missing values with medians
    X_test[i].fillna(
        X_train[i].median(), inplace=True
    )  # we have to use the train set to avoid leaking info about test set to the model.
    X_train[i].fillna(
        X_train[i].median(), inplace=True
    )  # have to fill in X_train after filling in X_test, otherwise median values considered in median calculation

    ## Normalise data (put between 0 and 1)
    X_test[i] = (X_test[i] - X_train[i].min()) / (
        X_train[i].max() - X_train[i].min()
    )  # min/max scaling
    X_train[i] = (X_train[i] - X_train[i].min()) / (
        X_train[i].max() - X_train[i].min()
    )  # min/max scaling

print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

## Classification Model and Evaluation

```python
from sklearn.ensemble import (
    AdaBoostClassifier,
    GradientBoostingClassifier,
    RandomForestClassifier,
)
from sklearn.linear_model import LogisticRegression
```

```python
model_reg = LogisticRegression(max_iter=10000)
model_reg = model_reg.fit(X_train, y_train)

preds_reg = model_reg.predict(X_test)
```

```python
model_rf = RandomForestClassifier(
    n_estimators=10, max_depth=None, min_samples_split=2, random_state=0
)
model_rf.fit(X_train, y_train)

preds_rf = model_rf.predict(X_test)
```

```python
model_ada = AdaBoostClassifier(n_estimators=15)
model_ada.fit(X_train, y_train)

preds_ada = model_ada.predict(X_test)
```

```python
model_grad = GradientBoostingClassifier(
    loss="log_loss", n_estimators=10, learning_rate=1.0, max_depth=1, random_state=0
)
model_grad.fit(X_train, y_train)

preds_grad = model_grad.predict(X_test)
```

As long as it predicts positive, then that's a win.
Don't want prediciting positive when negative.

```python
print(
    "Blind Invest: ", pct_change[test_idx].mean()
)  # on average, how much was the return (if I invested blindly and evenly across all S&P500)
print(
    "ML Invest (Logistic Regression): ",
    pct_change[y_test[preds_reg > 0].index.values].mean(),
)  # take true y_test values where profit prediciton, then average across all percent changes
print(
    "ML Invest (Random Forest): ",
    pct_change[y_test[preds_rf > 0].index.values].mean()
)
print("ML Invest (AdaBoost): ",
      pct_change[y_test[preds_ada > 0].index.values].mean()
)
print("ML Invest (Gradient Boosting): ",
      pct_change[y_test[preds_grad > 0].index.values].mean()
)
```
